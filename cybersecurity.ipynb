```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Threat_Level', order=df['Threat_Level'].value_counts().index, palette='coolwarm')
plt.title('Threat Level Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(y=df['Attack_Type'], order=df['Attack_Type'].value_counts().index, palette='viridis')
plt.title('Most Common Attack Types')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Threat_Level', y='Response_Time', palette='coolwarm')
plt.title('Response Time by Threat Level')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(y=df['Affected_System'], order=df['Affected_System'].value_counts().index, palette='Blues')
plt.title('Most Attacked System Types')
plt.show()
```

```python
df['Security_Team'].value_counts().plot(kind='bar', color='purple')
plt.title('Incidents per Security Team')
plt.show()
```

```python
avg_response = df.groupby('Attack_Type')['Response_Time'].mean().sort_values()
avg_response.plot(kind='barh', color='red')
plt.title('Average Response Time by Attack Type')
plt.show()
```

```python
df['Data_Compromised'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Data Compromised Proportion')
plt.show()
```

```python
sns.countplot(data=df, x='Status', palette='pastel')
plt.title('Incident Status Distribution')
plt.show()
```

```python
top_ips = df['IP_Address'].value_counts().nlargest(10)
top_ips.plot(kind='bar', color='orange')
plt.title('Top 10 Frequent IP Addresses')
plt.show()
```

```python
df['Threat_Numeric'] = df['Threat_Level'].map({'Low': 1, 'High': 2, 'Critical': 3})
plt.figure(figsize=(6, 4))
sns.heatmap(df[['Response_Time', 'Threat_Numeric']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Response Time and Threat Level')
plt.show()
```

```python
df.set_index('Timestamp').resample('D').size().plot(title='Incidents Per Day', color='blue')
plt.show()
```

```python
critical_percent = (df['Threat_Level'] == 'Critical').mean() * 100
print(f'Critical threats percentage: {critical_percent:.2f}%')
```

```python
avg_team_response = df.groupby('Security_Team')['Response_Time'].mean()
avg_team_response.plot(kind='bar', color='cyan')
plt.title('Average Response Time by Security Team')
plt.show()
```

```python
df['Month'] = df['Timestamp'].dt.month
df['Month'].value_counts().sort_index().plot(kind='bar', color='magenta')
plt.title('Incidents Per Month')
plt.show()
```
